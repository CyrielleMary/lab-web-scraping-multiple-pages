Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page: url ='https://en.wikipedia.org/wiki/Python'

```python
url = "https://en.wikipedia.org/wiki/Python"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select("a")
```

```python
for l in soup.select("a"):
    print(l["href"])
```

List all language names and number of related articles in the order they appear in wikipedia.org: url = 'https://www.wikipedia.org/'

```python
url = 'https://www.wikipedia.org/'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select('ul')
```

```python
soup.select('ul > li > a ')
```

```python
soup.select('mp-bottom > div:nth-child(6) > div > ul > li:nth-child(1) > div.hlist.wikipedia-languages-langs.inline > ul > li:nth-child(9) > a ')
```

```python
for a in soup.select("ul > li > a"):
    abr = a.get("lang")
    lang = a.get("title")
    print('abreviation: ', abr, 'language: ',lang)
```

A list with the different kind of datasets available in data.gov.uk: url = 'https://data.gov.uk/'

```python
url = 'https://ckan.publishing.service.gov.uk/dataset/'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select('#content > div.row.wrapper > aside > div > div > section:nth-child(1) > nav > ul')
```

```python
dataset = soup.select('#content > div.row.wrapper > aside > div > div > section:nth-child(1) > nav > ul')
for a in enumerate(dataset):
    label = a.select("span", class_="item-label").text.strip()
    print(label)
```
